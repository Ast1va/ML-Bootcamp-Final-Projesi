```python
# Imports
from pathlib import Path

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
# Dataset path, load, shape, and first rows
data_path = Path("../../data/raw/online_shoppers_intention.csv")
df = pd.read_csv(data_path)
print(f"Loaded shape: {df.shape}")
df.head()
```

```python
# Data info
df.info()
```

```python
# Basic describe (all columns)
df.describe(include='all')
```

```python
# Missing values analysis
missing = df.isna().sum().to_frame(name="missing_count")
missing
```

```python
# Target (Revenue) distribution counts and plot
target_col = 'Revenue'
revenue_counts = df[target_col].value_counts(dropna=False)
revenue_norm = df[target_col].value_counts(normalize=True, dropna=False)
print('Counts:\n', revenue_counts)
print('\nNormalized (%):\n', (revenue_norm * 100).round(2))
plt.figure(figsize=(5, 4))
ax = sns.countplot(x=target_col, data=df)
ax.bar_label(ax.containers[0])
plt.title('Revenue Distribution')
plt.tight_layout()
plt.show()
```

```python
# Numeric columns distribution (histograms)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(f'Numeric columns ({len(num_cols)}):', num_cols)
if num_cols:
    df[num_cols].hist(figsize=(14, 10), bins=20, layout=(-1, 3))
    plt.tight_layout()
    plt.show()
else:
    print('No numeric columns detected.')
```

```python
# Categorical columns distribution (countplots)
cat_cols = df.select_dtypes(include=['object', 'bool']).columns.tolist()
print(f'Categorical columns ({len(cat_cols)}):', cat_cols)
if cat_cols:
    for col in cat_cols:
        plt.figure(figsize=(6, 4))
        ax = sns.countplot(y=col, data=df, order=df[col].value_counts().index)
        ax.bar_label(ax.containers[0])
        plt.title(f'Distribution of {col}')
        plt.tight_layout()
        plt.show()
else:
    print('No categorical columns detected.')
```

```python
# Correlation matrix (Revenue cast to int)
df_corr = df.copy()
df_corr['Revenue_int'] = df_corr['Revenue'].astype(int)
corr_cols = df_corr.select_dtypes(include=['int64', 'float64']).columns.tolist()
corr = df_corr[corr_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=False, cmap='coolwarm', center=0, square=True)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
